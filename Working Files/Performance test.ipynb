```python
# Add Matplotlib inline magic command
%matplotlib inline

# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to check these)
starr_data_to_load = "starr_performance_data.csv"

# Read the City and Ride Data
starr_data_df = pd.read_csv(starr_data_to_load)
```

```python
# Display the data table for preview
starr_data_df.head()
```

```python
# Find missing values
starr_data_df.isnull().sum()
```

```python
# replace the NaNs with 0
starr_data_df = starr_data_df.fillna(0)
starr_data_df
```

```python
# remove symbols 
cols = starr_data_df.drop(starr_data_df.loc[:, 'State':'EL (Current & Monitored)'] axis=1)
starr_data_df[cols] = starr_data_df[cols].str.replace('%', "")
starr_data_df.dtypes
```

```python
# convert text data to numerical data type
cols = starr_data_df.columns.drop('Grade', 'Performance level', 'Subject', 'Year')
starr_data_df[cols] = starr_data_df[cols].apply(pd.to_numeric, errors='coerce')
starr_data_df.dtypes
```
